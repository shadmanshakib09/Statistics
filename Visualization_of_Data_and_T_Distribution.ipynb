```python
#box plot
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='species', y='sepal_length')
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Create subplots with 2 rows and 2 columns (one plot for each feature)
fig, axes = plt.subplots(2, 2, figsize=(14, 12))  # Adjusted figsize for clarity

# Q-Q Plot for Sepal Length
sm.qqplot(df['sepal_length'], line='45', ax=axes[0, 0])
axes[0, 0].set_title('Q-Q Plot of Sepal Length')
axes[0, 0].set_xlabel('Theoretical Quantiles')
axes[0, 0].set_ylabel('Sample Quantiles')

# Q-Q Plot for Sepal Width
sm.qqplot(df['sepal_width'], line='45', ax=axes[0, 1])
axes[0, 1].set_title('Q-Q Plot of Sepal Width')
axes[0, 1].set_xlabel('Theoretical Quantiles')
axes[0, 1].set_ylabel('Sample Quantiles')

# Q-Q Plot for Petal Length
sm.qqplot(df['petal_length'], line='45', ax=axes[1, 0])
axes[1, 0].set_title('Q-Q Plot of Petal Length')
axes[1, 0].set_xlabel('Theoretical Quantiles')
axes[1, 0].set_ylabel('Sample Quantiles')

# Q-Q Plot for Petal Width
sm.qqplot(df['petal_width'], line='45', ax=axes[1, 1])
axes[1, 1].set_title('Q-Q Plot of Petal Width')
axes[1, 1].set_xlabel('Theoretical Quantiles')
axes[1, 1].set_ylabel('Sample Quantiles')

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
#violon plot
plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='species', y='sepal_length',inner = 'quartile')
plt.title('Violin Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.pairplot(data=df, hue='species', markers=['o','s','D'])
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

```python
samples = []
stds = []
for i in range(10):
    x = df['sepal_length'].sample(15).values
    stds.append(x.std())
    samples.append(x.tolist())
```

```python
np.array(stds)
```

```python
samples = np.array(samples)
samples.shape
```

```python
sampling_means = samples.mean(axis=1)
sampling_means
```

```python
sample_std = np.mean(stds)
sample_std
```

```python
sns.distplot(sampling_means)
```

```python
sampling_means.mean()
```

```python
sampling_means.std()
```

```python
sampling_means.std()/np.sqrt(15)
```

```python
#95% confidence level and 15 sample
lower_limit = sampling_means.mean() - 2.131*sample_std/np.sqrt(15)
upper_limit = sampling_means.mean() + 2.131*sample_std/np.sqrt(15)
```

```python
print(lower_limit,upper_limit)
```

**We are 95% confident that the true mean sepal length for the population of Iris flowers lies between 5.44 and 6.248 cm.**
