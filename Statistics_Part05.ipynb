# Binomial Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
n = 10
p = 0.5
size = 1000
data = np.random.binomial(n, p, size)
plt.hist(data)
```

```python
sns.distplot(data)
```

```python
#if we increase the probability of success
n1 = 10
p1= 0.8
size = 1000
data1 = np.random.binomial(n1, p1, size)
plt.hist(data1)
```

```python
sns.distplot(data1) #graph will shift to right side
```

```python
#if we reduce the prob. of success
n2 = 10
p2 = 0.1
size = 1000
data2 = np.random.binomial(n2, p2, size)
plt.hist(data2)
```

```python
sns.distplot(data2)
```

```python
sns.distplot(data,label='P=0.5')
sns.distplot(data1,label='P=0.8')
sns.distplot(data2,label='P=0.1')
plt.legend()
```

# Center Limit Theorem

```python
sample_size = 30
num_of_sample = 1000
distribution_range = (0,1)

samples= np.random.uniform(distribution_range[0],distribution_range[1], size=(num_of_sample,sample_size))
samples_mean = np.mean(samples,axis=1)
plt.hist(samples_mean)
```

```python
sns.distplot(samples_mean)
```

```python
sample_size = 30
num_of_sample = 1000
distribution_range = (0,1)
theoretical_mean = (distribution_range[1]+distribution_range[0])/2
theoretical_var= (distribution_range[1]-distribution_range[0])**2/(12)
print("Mean before Sampling distribution:", theoretical_mean)
print("Var before Sampling distribution:",theoretical_var)
samples= np.random.uniform(distribution_range[0],distribution_range[1], size=(num_of_sample,sample_size))
samples_mean = np.mean(samples,axis=1)
print("Mean after Sampling distribution:", np.mean(samples_mean))
print("Var after Sampling distribution:", np.var(samples_mean))
# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot the original samples distribution
sns.histplot(samples.flatten(), bins=30, ax=axes[0], kde=True, color='blue')
axes[0].set_title('Original Uniform Distribution Samples')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

# Plot the distribution of sample means
sns.histplot(samples_mean, bins=30, ax=axes[1], kde=True, color='orange')
axes[1].set_title('Distribution of Sample Means')
axes[1].set_xlabel('Sample Mean')
axes[1].set_ylabel('Frequency')

# Display the plots
plt.tight_layout()
plt.show()
#see, the mean is same before and after the center limit theorem and the value for variance is 0.08/30
```

```python
0.08/30
```

```python
sns.distplot(samples_mean,label='Sample Mean')
sns.distplot(samples, label = 'Sample')
plt.legend()
```

# Lets practise Center Limit Theorem using iris dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.distplot(iris['petal_length']) #assuming this is our population data
```

```python
#Check this is normal disrribution or not(QQ plot)
import statsmodels.api as sm
sm.qqplot(iris['petal_length'], line='45')
plt.show()
```

```python
iris['petal_length'].mean()
```

```python
iris['petal_length'].std()
```

```python
#now divide the data into samples
samples = []
for i in range(100):
    samples.append(iris['petal_length'].sample(50).values.tolist())
```

```python
samples = np.array(samples)
```

```python
sample_means = samples.mean(axis=1)
```

```python
sns.distplot(sample_means)
```

```python
sample_means.mean()
```

```python
sample_means.std()/np.sqrt(50)
```

```python
lower_limit = sample_means.mean() - 2*sample_means.std()
upper_limit = sample_means.mean() + 2*sample_means.std()
print(lower_limit, upper_limit)
```
